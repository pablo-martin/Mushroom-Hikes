## Downloading Images from Server

Let's try downloading all the pictures of a given species

```python
#sample mushroom
my_name_id = 4
image_names = obs[obs.name_id == my_name_id].thumb_image_id.values
def generate_url_image_API(image_names):
    return ['https://images.mushroomobserver.org/960/' + str(int(w)) + '.jpg' \
                     for w in image_names if not np.isnan(w)]
```

```python
import requests

def download_pictures(url_addresses):
    for i, url in enumerate(url_addresses):
        r = requests.get(url, allow_redirects=True)
        open('../pics/test' + str(i) + '.jpg', 'wb').write(r.content)
    return
```

## Finalizing Image Labels (Species Names)
Let's check out what names are included, and see how much variability there is in naming conventions. This is where labels will come from so we have to be very careful.

```python
print names.shape
display(names)

def text_name_to_gs(text_name):
    fine_labels = {}
    split_name = text_name.split(' ')
    if len(split_name) == 1:
        fine_labels['Genus'] = split_name[0]
    elif len(split_name) == 2:
        fine_labels['Genus'] = split_name[0]
        fine_labels['Species'] = split_name[1]
    elif len(split_name) == 3:
        if split_name[2] == 'group':
            fine_labels['Genus'] = split_name[0]
            fine_labels['Species'] = split_name[1]     
        elif split_name[1] == 'subgenus':
            #technically not correct
            fine_labels['Genus'] = split_name[0]
            fine_labels['Species'] = split_name[2] 
    elif len(split_name) > 3:
        fine_labels['Genus'] = split_name[0]
        if split_name[1] != 'subgenus':
            fine_labels['Species'] = split_name[1]  
        else:
            fine_labels['Species'] = split_name[2]
    return fine_labels
```

```python

#extracts taxa information that it can find from the string
labels = ['Domain','Kingdom','Phylum','Class','Order','Family']

def extract_taxa(taxa_str):
    taxa_dict = {}
    for label in labels:
        sp = taxa_str.find(label)
        if sp >= 0:
            name_sp = sp + taxa_str[sp:].find('_') + 1
            name_ep = name_sp + taxa_str[name_sp + 1:].find('_') + 1
            taxa_dict[label] = taxa_str[name_sp:name_ep]
    return taxa_dict

#given weird csv, it compiles all the taxa information available there
def retrieve_taxa(observations):
    taxa = {}
    classified = \
        observations.classification.apply(lambda x: isinstance(x, str))
    for i,w in enumerate(observations[classified].index):
        name_id = int(observations.loc[w].name_id)
        base = observations.loc[w].classification
        v = 1
        while not unicode(observations.loc[w+v].id).isnumeric():
            base += observations.loc[w+v].id
            v += 1
        taxa[name_id] = extract_taxa(base)
        if i%10000:
            print('finished %i/%i observation' 
                            %(i,len(observations[classified].index)))
    return taxa
```

```python
#this takes a really long time
taxa = retrieve_taxa(observations)
```

## Filling in Genus and Species Pairs
Let's link up the Genus and species information with the taxa as well. 

```python
def is_genus_species(str):
    tmp = str.split(' ')
    if len(tmp) == 2:
        return True
    else:
        return False
    
tmp = sum(names.text_name.apply(lambda x: is_genus_species(x)))  
print('number of genus-species pair in names list: %i/%i'
                                             %(tmp, len(names)))
```

```python
def resolve_synonyms(id):
    if not np.isnan(names[names.id==id].synonym_id.values[0]):
        synonym_id = int(names[names.id==id].synonym_id.values[0])
        tmp = names[names.synonym_id == synonym_id]
        tmp = tmp[tmp.deprecated == 0]
        if tmp.shape[0] > 0:
            return tmp.iloc[np.random.choice(tmp.shape[0])]['text_name']
        else:
            return names[names.id==id].text_name.values[0]
    else:
        return names[names.id==id]['text_name'].values[0]
    
```

```python
for i, name_id in enumerate(taxa.keys()):
    text_name = resolve_synonyms(name_id)
    fine_labels = text_name_to_gs(text_name)
    for key, val in fine_labels.iteritems():
        taxa[name_id][key] = val
    if i%250 == 0:
        print('done with %i/%i species' %(i,len(taxa.keys())))
```

```python
pickle.dump(taxa, open(table_dir + 'taxa_dict.p', 'wb'))
```

```python
columns = ['id','location_id','in_location','vote_cache','when',
           'Domain', 'Kingdom', 'Phylum', 'Class', 'Order', 'Family',
           'Genus', 'Species', 'Image', 'Data_Source']

def print_elapsed_time():
    start = time.time()
    print('t: %1.4fms' %((time.time() - start) * 1000))
    return
    
import time
def create_df_from_observation(cid):
    start = time.time()
    
    tmp_obs = obs[obs.id == cid]
    in_taxa = int(tmp_obs.name_id) in taxa.keys()
    

    print_elapsed_time()
    
    cimages = images[images.observation_id == cid]
    df = pd.DataFrame(np.zeros([cimages.shape[0], len(columns)]),
                                              columns=columns)

    print_elapsed_time()
    
    df.loc[:,'id'] = cid
    df.loc[:,'location_id'] = int(tmp_obs.location_id)
    df.loc[:,'in_location'] = int(tmp_obs.is_collection_location)
    df.loc[:,'vote_cache'] = np.float(tmp_obs.vote_cache)
    df.loc[:,'when'] = str(tmp_obs.when.values)
    df.loc[:,'Image'] = cimages.image_id.values
    df.loc[:,'Data_Source'] = 'MO'

    print_elapsed_time()
    
    if in_taxa:
        #taxa information
        for taxa_subfield in ['Domain','Kingdom','Phylum',
                              'Class','Order','Family','Genus','Species']:
            df.loc[:,taxa_subfield] = \
                            taxa[int(tmp_obs.name_id)].get(taxa_subfield)
    else:
        print('not in taxa')
        GS_dict = text_name_to_gs(names[names.id == int(tmp_obs.name_id)].text_name.values[0])
        df.loc[:, 'Genus'] = GS_dict['Genus']
        df.loc[:, 'Species'] = GS_dict['Species']
        
    print_elapsed_time()
    return df

for cid in range(6,9):
    create_df_from_observation(cid)
#print((obs.shape[0] * 17.) / (1000 * 60 * 60))
```

```python
comp_df = []
for i, cid in enumerate(obs.id):
    if i%1000 == 0:
        print('on observation %i/%i' %(i,obs.shape[0]))
    tmp = create_df_from_observation(cid)
    if not None:
        comp_df.append(tmp)
        
df = pd.concat(comp_df)
df.loc[:,'location_id'] = df.location_id.astype('int32')
df.loc[:,'in_location'] = df.in_location.astype('int16')

display(df)
```
